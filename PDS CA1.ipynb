###### Questions we have about the data:

##### 1.1 Average Daily Volume Entering the City

```python
import warnings 
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
warnings.filterwarnings('ignore')

data_avgVolume=np.genfromtxt("average-daily-traffic-volume-entering-the-city.csv",
                names=True,
                dtype=[int,int],
                delimiter=',')

print(data_avgVolume)
print(data_avgVolume.dtype)
```

```python
years=np.unique(data_avgVolume["year"])
years

avgDailyV=data_avgVolume["ave_daily_traffic_volume_entering_city"]
avgDailyV
```

##### 1.2 Average Speed during Peak Hour 

```python
import warnings 
import numpy as np 
import matplotlib.pyplot as plt

data_avgSpeed=np.genfromtxt("average-speed-during-peak-hours.csv",
              dtype=[int,float,float],
              names=True,
              delimiter=",")

print(data_avgSpeed)
print(data_avgSpeed.dtype)
```

```python
express=data_avgSpeed["ave_speed_expressway"]
arterial=data_avgSpeed["ave_speed_arterial_roads"]
```

```python
## Output of Code 

title="Road Traffic Conditions during Peak Hours"
titlelen=len(title)
print("{:*^{titlelen}}".format(title,titlelen=titlelen+6))
print()

print("There are {} rows in Average Daily Volume Entering City dataset and {} rows for Average Speed during Peak Hours dataset.".format(len(data_avgVolume),len(data_avgSpeed)))
print("There are {} years of data which are from {} to {} in Average Daily Volume Entering City dataset.".format((len(years)),(years[0]),(years[-1])))
print()

set_year1=set(data_avgVolume['year'])
set_avg1=set(data_avgVolume['ave_daily_traffic_volume_entering_city'])
print("For Average Volume Entering City dataset, there are " + str(len(set_year1)) + " unique values in 'year' column and " + str(len(set_avg1 )) + " unique values in 'average daily traffic volume' column.")
print()

set_year2=set(data_avgSpeed['year'])
set_avg2=set(data_avgSpeed['ave_speed_expressway'])
set_avg3=set(data_avgSpeed['ave_speed_arterial_roads'])
print("For Average Speed during Peak Hours dataset, there are " + str(len(set_year2)) + " unique values in 'year' column and " + str(len(set_avg2)) + " unique values in 'average speed on expressway' column and "+ str(len(set_avg3)) + " for average speed on 'arterial roads' column.")
print()

min_value=data_avgVolume['ave_daily_traffic_volume_entering_city'].min()
argmin_value=data_avgVolume['ave_daily_traffic_volume_entering_city'].argmin()
max_value=data_avgVolume['ave_daily_traffic_volume_entering_city'].max()
argmax_value=data_avgVolume['ave_daily_traffic_volume_entering_city'].argmax()
print("The lowest number of traffic volume entering the city was in the year {} for with {}.".format(data_avgVolume[argmin_value]['year'],data_avgVolume[argmin_value]['ave_daily_traffic_volume_entering_city']))
print("The highest number of traffic volume entering the city was in the year {} for with {}.".format(data_avgVolume[argmax_value]['year'],data_avgVolume[argmax_value]['ave_daily_traffic_volume_entering_city']))  
print()

print("The mean for 'daily traffic volume entering the city' over the years is {:.0f}.".format(np.mean(data_avgVolume['ave_daily_traffic_volume_entering_city'])))
print("The mean for 'speed of traffic during peak hour' on expressway over the years is {:.0f}km/h.".format(np.mean(data_avgSpeed['ave_speed_expressway'])))
print("The mean for 'speed of traffic during peak hour' on arterial road over the years is {:.0f}km/h.".format(np.mean(data_avgSpeed["ave_speed_arterial_roads"])))
print()

print("The median for 'daily traffic volume entering the city' over the years is {:.0f}.".format(np.median(data_avgVolume['ave_daily_traffic_volume_entering_city'])))
print("The median for 'speed of traffic during peak hour' on expressway over the years is {:.0f}km/h.".format(np.median(data_avgSpeed['ave_speed_expressway'])))
print("The median for 'speed of traffic during peak hour' on arterial road over the years is {:.0f}km/h.".format(np.median(data_avgSpeed["ave_speed_arterial_roads"])))
print()

var=np.var(data_avgVolume['ave_daily_traffic_volume_entering_city'])
sd=np.std(data_avgVolume['ave_daily_traffic_volume_entering_city']) 
print("The variance for 'Average Daily Traffic Entering City' is {:.0f}.".format(var))
print("The standard deviation for 'Average Daily Traffic Entering City' is {:.0f}.".format(sd))
print()

varExpress=np.var(data_avgSpeed['ave_speed_expressway'])
varAr=np.var(data_avgSpeed['ave_speed_arterial_roads'])
sdExpress=np.std(data_avgSpeed['ave_speed_expressway'])
sdAr=np.std(data_avgSpeed['ave_speed_arterial_roads'])
print("The variance for average speed on express way and arterial roads are {:.2f} and {:.2f}.".format((varExpress),(varAr)))
print("The standard deviation for average speed on express way and arterial roads are {:.2f} and {:.2f}.".format(sdExpress,sdAr))
```

```python
plt.figure(figsize=(9,6))
plt.plot(years,avgDailyV,"r--")
plt.title("Average Daily Traffic Volume Entering the City",fontsize=14)
plt.xlabel("YEAR")
plt.ylabel("AVERAGE TRAFFIC VOLUME")
plt.show()
```

```python
plt.figure(figsize=(9,6))
plt.scatter(arterial,avgDailyV,s=200,color="slateblue")

#line of best fit 
plt.plot(np.unique(arterial), np.poly1d(np.polyfit(arterial, avgDailyV, 1))(np.unique(arterial)))
plt.title("Scatter Plot for Average Daily Traffic Volume versus Average Speed during Peak Hours",fontsize=14)
plt.xlabel("AVERAGE SPEED ON ARTERIAL ROADS (KM/H)")
plt.ylabel("AVERAGE TRAFFIC VOLUME")

plt.show()
```

```python
plt.figure(figsize=(9,6))
plt.scatter(express,avgDailyV,s=200,color="mediumaquamarine")

#line of best fit 
plt.plot(np.unique(express), np.poly1d(np.polyfit(express, avgDailyV, 1))(np.unique(express)))
plt.title("Scatter Plot for Average Daily Traffic Volume versus Average Speed during Peak Hours",fontsize=14)
plt.xlabel("AVERAGE SPEED ON EXPRESSWAY (KM/H)")
plt.ylabel("AVERAGE TRAFFIC VOLUME")

plt.show()
```

### Dataset 2
## Number of Traffic Lights 

###### Questions we have about the data: 
1. What is the trend in number of traffic lights over the years? 
2. Which year has the highest number of traffic lights? 
3. What is the average number of traffic lights over the years? 

```python
import warnings 
import numpy as np 
import matplotlib.pyplot as plt

data_tl = np.genfromtxt("traffic-lights.csv",
                      dtype=(int,int),
                      delimiter=",",
                      names=True)
data_tl
```

```python
years_tl=np.unique(data_tl["year"])
years_tl
```

```python
tl=data_tl["traffic_lights"]
tl
```

```python
## Output of Code 

title2="Number of Traffic Lights"
title2_len=len(title2)
print("{:*^{title2_len}}".format(title2,title2_len=title2_len+6))
print()

print("There are {} rows in this dataset".format(len(data_tl)))
print("There are {} years of data which are from {} to {}.".format((len(years_tl)),(years_tl[0]),(years_tl[-1])))
print()

set_y2=set(data_tl['year'])
set_tl=set(data_tl['traffic_lights'])
print("There are " + str(len(set_y2)) + " unique values in year column.")
print("There are " + str(len(set_tl)) + " unique values in traffic_lights column.")
print()

min_tl=data_tl['traffic_lights'].min()
argmin_tl=data_tl['traffic_lights'].argmin()
max_tl=data_tl['traffic_lights'].max()
argmax_tl=data_tl['traffic_lights'].argmax()
print("The lowest number of traffic lights was in the year {} with {}.".format(data_tl[argmin_tl]['year'],data_tl[argmin_tl]['traffic_lights']))
print("The highest number of traffic lights was in the year {} with {}.".format(data_tl[argmax_tl]['year'],data_tl[argmax_tl]['traffic_lights']))
print()

var_tl=np.var(data_tl['traffic_lights'])
sd_tl=np.std(data_tl['traffic_lights'])
print("The mean of number of traffic lights over the years is {:.2f}".format(np.mean(data_tl['traffic_lights'])))
print("The median of the number of traffic lights over the years is {:.0f}.".format(np.median(data_tl['traffic_lights'])))
print("The variance of number of traffic lights is {:.2f}".format(var_tl))
print("The standard devition of number of traffic lights is {:.2f}".format(sd_tl))
```

```python
plt.figure(figsize=(12, 6))
plt.bar(years_tl,tl,width=0.8,color='cadetblue')
plt.title("Number of Traffic Lights",fontsize=14)
plt.xlabel("YEARS")
plt.ylabel("NUMBER OF TRAFFIC LIGHTS")

for a,b in zip(years_tl, tl):
    plt.text(a, b, str(b))

plt.tight_layout()
plt.show()
```

### Dataset 3
## Annual Mileage for Private Motor Vehicles
##### Average Annual Kilometers Travelled per Vehicles 

###### Questions we have about the data: 
1. What is the observed trends in annual mileage for private motor vehicles over the years? 
2. Which vehicle has the highest mileage? 
3. Which vehicle has the lowest mileage? 
4. What is the average mileage for each vehicle over the years? 

```python
import warnings 
import numpy as np 
import matplotlib.pyplot as plt

data_AM = np.genfromtxt("average-annual-kilometres-travelled-per-vehicle.csv",
                      dtype=(int,"U100",int),
                      delimiter=",",
                      names=True)
print(data_AM)
print(data_AM.dtype)
```

```python
yearAM=np.unique(data_AM["year"])
yearAM
```

```python
PriM=np.unique(data_AM["vehicle_type"])
PriM
```

```python
car=data_AM[data_AM["vehicle_type"]=="Cars"]["average_annual_mileage"]
heavy=data_AM[data_AM["vehicle_type"]=="Heavy Goods Vehicle"]["average_annual_mileage"]
light=data_AM[data_AM["vehicle_type"]=="Light Goods Vehicle"]["average_annual_mileage"]
motor=data_AM[data_AM["vehicle_type"]=="Motorcycles"]["average_annual_mileage"]
hBus=data_AM[data_AM["vehicle_type"]=="Private Hire Buses"]["average_annual_mileage"]
sBus=data_AM[data_AM["vehicle_type"]=="School Buses"]["average_annual_mileage"]

data_plot=[car,heavy,light,motor,hBus,sBus]

print(car,heavy,light,motor,hBus,sBus)
```

```python
avgMileage=data_AM["average_annual_mileage"]
avgMileage
```

```python
## Output of Code 

title3="Annual Mileage for Private Motor Vehicles"
title3_len=len(title3)
print("{:*^{title3_len}}".format(title3,title3_len=title3_len+6))
print()

print("There are {} rows in this dataset.".format(len(data_AM)))
print("There are {} vehical classes in this dataset.".format(len(PriM)))

print("There are {} years of data which are from {} to {}.".format((len(yearAM)),(yearAM[0]),(yearAM[-1])))
print()

set_y3=set(data_AM['year'])
set_vt=set(data_AM['vehicle_type'])
set_avg=set(data_AM['average_annual_mileage'])
print("There are " + str(len(set_y3)) + " unique values in the year column.")
print("There are " + str(len(set_vt)) + " unique values in the vehicle types column.")
print("There are " + str(len(set_avg)) + " unique values in the average annual mileage column.")
print()

max_car=car.max()
min_car=car.min()
print("The lowest annual mileage for Car over the years is {}.".format(min_car))
print("The highest annual mileage for Car over the years is {}.".format(max_car))
print()

max_heavy=heavy.max()
min_heavy=heavy.min()
print("The lowest annual mileage for Heavy Goods Vehicle over the years is {}.".format(min_heavy))
print("The highest annual mileage for Heavy Goods Vehicle over the years is {}.".format(max_heavy))
print()

max_light=light.max()
min_light=light.min()
print("The lowest annual mileage for Light Goods Vehicle over the years is {}.".format(min_light))
print("The highest annual mileage for Light Goods Vehicle over the years is {}.".format(max_light))
print()

max_motor=motor.max()
min_motor=motor.min()
print("The lowest annual mileage for Motorcycles over the years is {}.".format(min_motor))
print("The highest annual mileage for Motorcycles over the years is {}.".format(max_motor))
print()

max_hBus=hBus.max()
min_hBus=hBus.min()
print("The lowest annual mileage for Private Hire Buses over the years is {}.".format(min_hBus))
print("The highest annual mileage for Private Hire Buses over the years is {}.".format(max_hBus))
print()

max_sBus=sBus.max()
min_sBus=sBus.min()
print("The lowest annual mileage for School Buses over the years is {}.".format(min_sBus))
print("The highest annual mileage for School Buses over the years is {}.".format(max_sBus))
print()

max_avg=data_AM['average_annual_mileage'].max()
argmax_avg_AM=data_AM['average_annual_mileage'].argmax()
min_avg=data_AM['average_annual_mileage'].min()
argmin_avg_AM=data_AM['average_annual_mileage'].argmin()
#argmax_type=data_AM['average_annual_mileage'].argmax()
##print(argmax_avg_AM)
##data_AM[argmax_avg_AM]['year']
##data_AM[argmax_avg_AM]['vehicle_type']

#data_AM.dtype
print("The maximum number of Annual Mileage for Private Motor Vehicles was in the year {} for {} with {}.".format(data_AM[argmax_avg_AM]['year'],data_AM[argmax_avg_AM]['vehicle_type'],data_AM[argmax_avg_AM]['average_annual_mileage']))
print("The minimum number of Annual Mileage for Private Motor Vehicles was in the year {} for {} with {}.".format(data_AM[argmin_avg_AM]['year'],data_AM[argmin_avg_AM]['vehicle_type'],data_AM[argmin_avg_AM]['average_annual_mileage']))
print()

var_AM=np.var(data_AM['average_annual_mileage'])
sd_AM=np.std(data_AM['average_annual_mileage'])
print("The mean of annual mileage for Private Motor Vehicles over the years is {:.2f}".format(np.mean(data_AM['average_annual_mileage'])))
print("The median of annual mileage for Private Motor Vehicles is {:.0f}".format(np.median(data_AM['average_annual_mileage'])))
print("The variance of annual mileage for Private Motor Vehicles is {:.0f}".format(var_AM))
print("The standard deviation of annual mileage for Private Motor Vehicles is {:.0f}".format(sd_AM))

```

```python
plt.figure(figsize=(12, 8))
p1=plt.plot(yearAM,car,"r-")
p2=plt.plot(yearAM,heavy,"y-")
p3=plt.plot(yearAM,light,"g-")
p4=plt.plot(yearAM,motor,"b-")
p5=plt.plot(yearAM,hBus,"m-")
p6=plt.plot(yearAM,sBus,"k-")
plt.title("Annual Mileage for Private Motor Vehicles",fontsize=14)
plt.xlabel("YEAR")
plt.ylabel("AVERAGE ANNUAL MILEAGE (KM)")
plt.legend((p1[0],p2[0],p3[0],p4[0],p5[0],p6[0]), ('Car','Heavy Goods Vehicle','Light Goods Vehicle','Motorcycles','Private Hire Buses','School Buses'))
plt.show()
```

```python
plt.figure(figsize=(12, 8))
bp = plt.boxplot(data_plot,labels=['Car','Heavy Goods Vehicle','Light Goods Vehicle','Motorcycles','Private Hire Buses','School Buses'],patch_artist=True)
colors=['lightgreen','pink','lightblue','hotpink','moccasin','lightsalmon']
patchcolor = list(zip(bp['boxes'], colors))

for patch,color in patchcolor:
    patch.set_facecolor(color)

for line in bp['medians']:
    x, y = line.get_xydata()[1] # top of median line
    # overlay median value
    plt.text(x, y, 'Median:{:.0f}'.format(y),
         horizontalalignment='center',fontsize=10) # draw above, centered

plt.title("Average Annual Mileage for Private Motor Vehicles",fontsize=14)
plt.xlabel("VEHICLE TYPE")
plt.ylabel("AVERAGE ANNUAL MILEAGE (KM)")
plt.show()
```
